```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
df["ans"] = iris["target"]
df
```

```python
from sklearn.cluster import KMeans
clu = KMeans(n_clusters=3)
clu.fit(iris["data"])
```

```python
df["cluster"] = clu.labels_
df
```

```python
# 圓心: 最標準東西的長相
clu.cluster_centers_
```

```python
import seaborn as sns
sns.scatterplot(x=df["sepal length (cm)"],
                y=df["petal length (cm)"],
               hue=clu.labels_)
```

```python
from sklearn.metrics import silhouette_score
ks = []
scores = []
for k in range(2, 10):
    clu = KMeans(n_clusters=k)
    clu.fit(iris["data"])
    ks.append(k)
    scores.append(silhouette_score(iris["data"], clu.labels_))
print(ks)
print(scores)
```

```python
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Sihouette Score v.s. K")
plt.xlabel("K")
plt.ylabel("score")
plt.plot(ks, scores, "p--m")
```
