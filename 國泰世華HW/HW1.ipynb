```python
import pandas as pd
import os
os.chdir("D:\Andy\Desktop\lvr_landcsv")
df_a = pd.read_csv("a_lvr_land_a.csv", encoding='UTF-8', low_memory=False)
df_b = pd.read_csv("b_lvr_land_a.csv", encoding='UTF-8', low_memory=False)
df_e = pd.read_csv("e_lvr_land_a.csv", encoding='UTF-8', low_memory=False)
df_f = pd.read_csv("f_lvr_land_a.csv", encoding='UTF-8', low_memory=False)
df_h = pd.read_csv("h_lvr_land_a.csv", encoding='UTF-8', low_memory=False)
```

```python
df_all = pd.concat([df_a,df_b,df_e,df_f,df_h],axis=0)
```

```python
df_all = df_all[df_all.duplicated()==False]  # 去掉重複的行(ex.英文翻譯)
# df_all.drop(axis=0, index=0, inplace=True)  
```

```python
filter1 = df_all["主要用途"] == "住家用"
filter2 = df_all["建物型態"] == "住宅大樓(11層含以上有電梯)"
df_filt1 = df_all[filter1 & filter2]
# 第三個條件碰到問題，因為沒有辦法直接比較中文數字大小
# 先過濾掉前兩個條件後在過濾第三個
```

```python
import numpy as np
total_floor = np.array(df_filt1["總樓層數"])
floor_list=total_floor.tolist()
floor_list
```

```python
def transform(s):
    num = 0
    dict1 = {'一': 1, '二': 2, '三': 3, '四': 4, '五': 5, '六': 6, '七': 7, '八': 8, '九': 9}
    if s: # 確保傳入的值恆為true
        idx_s = s.find('十')   # 找有沒有'十'
        if idx_s != -1:
            num += dict1.get(s[0], 1) * 10    # 十位數的值乘上10，沒有一十所以預設值為1
        if s[-1] in dict1:   # 個位數
            num += dict1[s[-1]]
    return num

# type(transform("三十四"))
```

```python
def transform2(l):
    list1 = []
    for x in l:
        x = str(x)
        if '層' in x:
            x = transform(x[:-1])
            if x >= 13:
                x = 'True'
        list1.append(x)  
    return list1
    
# transform2(floor_list)
```

```python
def unique_index(L, e):
    return [i for (i, j) in enumerate(L) if j == e]

floor_index = unique_index(transform2(floor_list), 'True')
floor_index
```

```python
df_filt2 = df_filt1.iloc[floor_index]
df_filt2.to_csv("filter_a.csv", encoding="big5", index=False)
```

```python
# df_all.shape 
# 總件數
def total_number(s):
    return s.shape[0]

total_number(df_all)
```

```python
cal_car = np.array(df_all["交易筆棟數"])
car_list=cal_car.tolist()
# car_list
```

```python
def total_cars(s):
    sum = 0
    for x in s:
        y = int(x[-1])   # 如果真的有兩位數車位，可以抓後三位然後判斷有沒有"車"
        sum += y   
    return sum

total_cars(car_list)   # 總車位數
```

```python
def avg_price(s):
    sum = 0
    for x in s:
        x = int(x)
        sum += x
    
    return sum / total_number(df_all)

avg_price(df_all["總價元"])  # 平均總價元
```

```python
def avg_price(s):
    sum = 0
    for x in s:
        x = int(x)
        sum += x
    
    return sum / total_cars(car_list)

avg_price(df_all["車位總價元"])  # 平均車位總價元
```

```python
data_b = {'總件數': [total_number(df_all)],
          '總車位數': [total_cars(car_list)],
          '平均總價元': [avg_price(df_all["總價元"])],
          '平均車位總價元': [avg_price(df_all["車位總價元"])]
         }    # value加中括號避免要給他index

df_filter_b = pd.DataFrame(data_b)
df_filter_b.to_csv("filter_b.csv", encoding="big5", index=False)
```
